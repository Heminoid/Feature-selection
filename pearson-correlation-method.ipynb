Our method uses the fact that the correlation of two independent random variables is 0

which means if a particular column is a randomly generated feature, its correlation with all the other features should be 0(not seen in practice as it requires infinite samples)

Since we have limited samples of all the RVs(4119), we calculate the pearson correlation coefficient between two RVs. We claim that over a large set of samples the pearson correlation is near 0(<0.05) with high probability(the proof follows from the law of large numbers)



```python
import seaborn as sns
import matplotlib.pyplot as plt
```

We now extract the pearson correlation between each feature and the remaining 39 features and the target y.

we will be noting how many of these correlations are <0.05, indicating a high chance of independence

```python

correlation_scores = []


correlation_threshold = 0.05


for feature in X.columns:
    # Calculating the Pearson correlation with all other features and the target
    correlations = df.drop(columns=[feature]).corrwith(df[feature], method='pearson')
    # Counting the number of correlations below the threshold
    below_threshold = (correlations.abs() < correlation_threshold).sum()
    
    correlation_scores.append(below_threshold)


scores_df = pd.DataFrame({
    'Feature': X.columns,
    'Correlation_Score': correlation_scores
})

# Plotting a bar chart using seaborn for visualisation
plt.figure(figsize=(12, 6))
sns.barplot(x='Correlation_Score', y='Feature', data=scores_df.sort_values(by='Correlation_Score', ascending=False))
plt.title('Correlation Scores of Features')
plt.xlabel('Correlation Score')
plt.ylabel('Feature')
plt.show()
```

```python
sorted_scores_df = scores_df.sort_values(by='Correlation_Score', ascending=False)
```

Lets have a look at how each variable ranks

Intuitively we will be picking the variables with the 20 lowest scores

```python
sorted_scores_df
```

Looks like the top 20 variables have a full score of 40 which indicates that its very likely that they are the independent randomly generated variables we were looking for!

```python
bottom_20_features = sorted_scores_df.tail(20)


bottom_20_features_list = bottom_20_features['Feature'].tolist()


print(bottom_20_features_list)
```

Generating final csv file!

```python
df_mark = pd.DataFrame({'Variables': X.columns, 'Mark': 0})

df_mark.loc[df_mark['Variables'].isin(bottom_20_features_list), 'Mark'] = 1

df_mark.to_csv('submission.csv', index=False)
```
